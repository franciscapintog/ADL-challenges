```python
def describe_frec(df):

  tmp = df

  for index, var in enumerate(tmp):
    if len(tmp[var].value_counts()) > 5:
      print(tmp[var].describe())
      print("\n\n")

    else:
      print(tmp[var].value_counts())
      print("\n\n")
```

```python
describe_frec(df_sample)
```

```python
def null_list(dataframe, var, print_list = False):

  tmp = dataframe
  null_vars = tmp[var].isnull().sum()

  print(f"Variable: {var}")
  print(f"Cantidad de casos perdidos: {null_vars}.")
  print(f"Porcentaje de casos perdidos: {round(null_vars/len(tmp) * 100, 2)} %")
  print("-------\n")

  if print_list == True:
    print(df_sample["wdi_exph"].isnull() == True)
```

```python
null_list(df, "wdi_exph", print_list = True)
```
