En cada caso, se incorporará primero la función mostrada en el enunciado, y posteriormente se hará una solución mejorada contemplando los requerimientos del Desafío.

```python
def create_random_row():
  
  age = random.randint(18, 90)
  income = random.randrange(10000, 1000000, step = 1000)
  employment_status = random.choice(["Unemployed", "Employed"])
  debt_status = random.choice(["Debt", "No debt"])
  churn_status = random.choice(["Churn", "No churn"])

  return age, income, employment_status, debt_status, churn_status
```

```python
random_database = [create_random_row() for i in range(10000000)]
```

**Ejercicio 4**

```python
churn_subset, no_churn_subset = [], []

for i in random_database:
  for j in i:
    if j == "Churn":
      churn_subset.append(i)

  for j in i:
    if j == "No churn":
      no_churn_subset.append(i)
```

1. <code>churn_subset</code> y <code>no_churn_subset</code> es una segmentación de <code>random_database</code> que separa a los clientes si se han cambiado de compañía o no (dentro de los supuestos del desafío).
2. La complejidad algorítmica del código es:

> $$Θ = (i \times j)^2$$

3. La implementación alternativa se encuentra en las siguientes celda de código. Se separarán los clientes que se han cambiado de compañía y a los que no con <code>filter</code>.
4. Los resultados del código entregado y de la solución propuesta son iguales.
5. La celda subsiguiente de código contiene la media, varianza, mínimo y máximo de edad de cada subset, considerando métodos nativos de Python.

Se utilizará una comprensión de lista para seleccionar las edades de cada subset, que finalmente serán sumadas y divididas por la cantidad de elementos para generar el promedio.

Posteriormente, se utilizará el cálculo anterior para obtener la varianza, restándolo de cada observación, y elevando esta resta al cuadrado. Esta operaicón se hará con el método <code>map</code> para aplicar la operación a cada observación, y posteriormente se utilizará <code>reduce</code> para sumar todos estos resultados, y dividirlos por la cantidad de elementos en la lista.

En el caso de los mínimos y máximos, se buscarán con los métodos <code>min</code> y <code>max</code>, nativos de Python.

```python
churn_subset2 = list(filter(lambda x: x[4] == "Churn", random_database))
no_churn_subset2 = list(filter(lambda x: x[4] == "No churn", random_database))
```

```python
churn_age = [x[0] for x in churn_subset2]
no_churn_age = [x[0] for x in no_churn_subset2]

churn_mean = (functools.reduce(lambda a,b : a + b, churn_age))/len(churn_age)
no_churn_mean = (functools.reduce(lambda a,b : a + b, no_churn_age))/len(no_churn_age)
```

```python
churn_variance = (functools.reduce(lambda a,b: a + b, list(map(lambda x: (x - churn_mean)**2, churn_age))))/len(churn_age)
no_churn_variance = (functools.reduce(lambda a,b: a + b, list(map(lambda x: (x - no_churn_mean)**2, no_churn_age))))/len(no_churn_age)
```

```python
churn_max = max(churn_age)
no_churn_max = max(no_churn_age)
```

```python
churn_min = min(churn_age)
no_churn_min = min(no_churn_age)
```

```python
print("Promedio de edad de clientes que se han cambiado de compañía: ", churn_mean, "\n",
      "Promedio de edad de clientes que no se han cambiado de compañía :", no_churn_mean, "\n",
      "Varianza de edad de clientes que se han cambiado de compañía :", churn_variance, "\n",
      "Varianza de edad de clientes que no se han cambiado de compañía :", no_churn_variance, "\n",
      "Edad máxima de clientes que se han cambiado de compañía: ", churn_max, "\n",
      "Edad máxima de clientes que no se han cambiado de compañía: ", no_churn_max, "\n",
      "Edad mínima de clientes que se han cambiado de compañía: ", churn_min, "\n",
      "Edad mínima de clientes que no se han cambiado de compañía: ", no_churn_min, "\n")
```
