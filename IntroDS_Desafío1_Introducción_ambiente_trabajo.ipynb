**Análisis DF primeros 15 resultados: DF3_1**

```python
import pandas as pd
import numpy as np

print(round(np.mean(df3_1["passengers"]), 1))
print("\n---\n")
print(round(np.median(df3_1["passengers"]), 1))
print("\n---\n")
print(round(np.std(df3_1["passengers"]), 1))
```

**Análisis DF últimos 15 resultados: DF3_2**

```python
import pandas as pd
import numpy as np

print(round(np.mean(df3_2["passengers"]), 1))
print("\n---\n")
print(round(np.median(df3_2["passengers"]), 1))
print("\n---\n")
print(round(np.std(df3_2["passengers"]), 1))
```

En relación a los resultados obtenidos, en el caso de los promedios este aumenta en un 104,6% (el promedio de las últimas 15 observaciones respecto al de las primeras 15), mostrando un aumento importante en el tráfico del periodo, lo que coincide además con el aumento de la mediana entre periodos en un 88% (igual comparación que en el caso anterior).

A pesar de lo anterior, es importante notar que la desviación estándar se octuplica en el segundo caso, lo que muestra una mayor oscilación entre la cantidad de pasajeros mes a mes, muy distinto al primer caso, donde la cantidad de pasajeros era mucho menor, pero también más estable en torno al promedio.
