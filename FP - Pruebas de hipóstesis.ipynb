Parte 2

```python
df["Africa_bin"] = np.where(df["region"] == "Africa", 1, 0)
df["America_bin"] = np.where(df["region"] == "Americas", 1, 0)
df["Asia_bin"] = np.where(df["region"] == "Asia", 1, 0)
df["Europe_bin"] = np.where(df["region"] == "Europe", 1, 0)
df["Oceania_bin"] = np.where(df["region"] == "Oceania", 1, 0)
```

```python
df
```

Parte 3

```python
def hip_test(df, var, binarize):

  df = df.dropna(subset = [var])
  tmp = df
  group0 = tmp[tmp[binarize] == 0][var]
  group1 = tmp[tmp[binarize] == 1][var]

  t, p_value = stats.ttest_ind(group0, group1)

  print(f"Media de group0: {np.mean(group0)}.")
  print(f"Media de group1: {np.mean(group1)}.")
  print(f"La diferencia entre las medias es de {np.mean(group0) - np.mean(group1)}.")
  print(f"Valor t de la prueba es: {t}.")
  print(f"Valor p de la prueba es: {p_value}.")
```

```python
hip_test(df, "femlab", "Africa_bin")
```

```python
hip_test(df, "femlab", "America_bin")
```

```python
hip_test(df, "femlab", "Asia_bin")
```

```python
hip_test(df, "femlab", "Europe_bin")
```

```python
hip_test(df, "femlab", "Oceania_bin")
```

Comentarios:

Nota: se analizarán los casos considerando un nivel de certeza de un 95% (ya que los resultados del método ttest_ind utilizan tal valor por defecto).

* Variable femlab:
1. En el caso de Africa, la media no es significativamente distinta de la del resto de continentes, dado que el p_value es mayor a 0.05, y el valor de t hace que el puntaje se encuentre dentro de la curva de distribución normal.
2. En el caso de América, la diferencia entre este continente y el resto de países no es significativa: nuevamente el estadístico de prueba p supera el valor de 0.05, y el estadístico t indica que el puntaje se encuentra dentro de la curva, no siendo posible rechazar la hipótesis nula.
3. En el caso de Asia es distinto: el estadístico p es menor a 0.05, y el estadístico t indica que el puntaje se encuentra fuera del 99% de la curva normal, por lo que la media de este continente en relación a la variable sí tiene diferencias estadísticamente significativas con el resto de territorios.
4. En el caso de Europa, ocurre lo mismo que en 3, a pesar de que está más cerca de los valores límite (diferencia del continente sí tiene diferencias estadísticamente significativas).
5. En el caso de Oceanía se vuelve a la situación vista en 1. y 2.: los estadísticos no superan el test de hipótesis, con lo que el nivel de la variable en este continente no tiene diferencias estadísticamente significativas con el resto del mundo.
