* **Desafío 3**

Parte I

```python
counter_div = 0
for i in array2:
  if i % 2 == 0 or i % 3 == 0:
    print(f"El número {round(i, 2)} es divisible por 2 o 3")
    counter_div += 1
  else:
    print(f"El número {round(i, 2)} no es divisible por 2 o 3")

print(f"\nLa cantidad de números del array divisibles por 2 o 3 es {counter_div}.")
```

Parte II

```python
counter_div = 0
for i in array2:
  if i % 2 == 0 and i % 3 == 0:
    print(f"El número {round(i, 2)} es divisible por 2 y 3")
    counter_div += 1
  else:
    print(f"El número {round(i, 2)} no es divisible por 2 y 3")

print(f"\nLa cantidad de números del array divisibles por 2 y 3 es {counter_div}.")
```

Parte III

*Nota: Para acortar impresión de resultados, solo se imprimen los resultados positivos del loop, y la conclusión solicitada.*

```python
counter_div = 0
for i in array2:
  if i % 3 == 0 and i % 2 != 0:
    print(f"El número {round(i, 2)} es divisible por 3 y no es divisible por 2")
    counter_div += 1
  else:
    pass

print(f"\nLa cantidad de números del array divisibles por 3 y no por 2 es {counter_div}.")
```

Parte IV

```python
counter_div = 0
for i in array2:
  if i % 2 != 0 and i % 3 != 0:
    print(f"El número {round(i, 2)} no es divisible por 2 ni por 3")
    counter_div += 1
  else:
    pass

print(f"\nLa cantidad de números del array no divisibles por 3 ni 2 es {counter_div}.")
```

* **Desafío 5**

Se inicia importando el DataFrame solicitado y creando la variable del promedio de pasajeros.

```python
df = pd.read_csv("/content/flights.csv")
```

```python
pass_mean = round(np.mean(df["passengers"]), 2)
print(f"El promedio de pasajeros del dataframe es de {pass_mean}.")
```

Se crea columna "underperforming" solo con valores 0 inicialmente

```python
df["underperforming"] = 0
```

Y posteriormente se itera para modificar los valores, respecto a la cantidad de pasajeros en relación al promedio.

```python
for index, row in df.iterrows():

  if row["passengers"] < pass_mean:
    df.at[index, "underperforming"] = 1
  else:
    df.at[index, "underperforming"] = 0
```

* **Desafío 6**

Se crea el valor de desviación estándar y se retoma el de promedio, ya obtenido para el Desafío 5.

```python
pass_std = round(np.std(df["passengers"]), 2)
print(f"La media de pasajeros es de {pass_mean}, y la desviación estándar es de {pass_std}.")
```

Se define "outlier", también inicialmente solo con valores 0. Posteriormente, se aplica un loop con la condición dada para modificar el valor de cada observación y para imprimir qué meses (par único año-mes) se escapan del rango, junto con realizar un conteo para saber la cantidad de observaciones que cumplen esta condición.

```python
df["outlier"] = 0

print(f"Los pares año-mes que tuvieron una cantidad de pasajeros fuera de rango son:\n")

for index, row in df.iterrows():
  
  if row["passengers"] < (pass_mean - pass_std) or row["passengers"] > (pass_mean + pass_std):
    df.at[index, "outlier"] = 1
    print(row["year"], row["month"])
  else:
    pass

outlier_vls = sum(df["outlier"])
print(f"\nLa cantidad de observaciones que superan el rango es de {outlier_vls}.")
```
