###**Desafío 5**

Parte I

1. Se parte creando la nueva columna "ha_ganado" con las condiciones dadas
2. Posteriormente, se crea la variable que agrupa la cantidad de países con uno o más partidos ganados por continente, para ver cuáles tienen partidos ganados
3. Luego, se calcula en una variable por continente, la cantidad de países con partidos ganados, dividido por la cantidad de países que integran el continente.

```python
df["ha_ganado"] = np.where(df["juegos_ganados"] != 0, 1, 0)
vic_continent = df.groupby("continent")["ha_ganado"].value_counts(normalize = True)
vic_continent
```

Conclusiones:
1. El continente con mayor probabilidad de tener países con al menos una victoria es Sudamérica, que no tiene países que no hayan ganado ningún partido, su probabilidad es 1.
2. El continente que presenta mayor similitud entre la probabilidad de tener países con al menos una victoria en comparación con no tener ninguna es África, que tiene un 0.6 de probabilidad en el primer caso, y por consiguiente, 0.4 en el segundo.
3. Adicionalmente, notamos que el continente asiático no tiene partidos ganados, por lo que su probabilidad de tener partidos ganados es 0.


```python
df["ha_perdido"] = np.where(df["juegos_perdidos"] != 0, 1, 0)
loss_continent = df.groupby("continent")["ha_perdido"].value_counts(normalize = True)
loss_continent
```

Conclusiones:
1. Los continentes con mayor probabilidad de tener países con al menos un partido perdido son África y Asia, que no tiene países que no hayan perdido ningún partido, su probabilidad es 1.
2. Los continentes que presentan mayor probabilidad de no tener países con uno o más partidos perdidos son Norteamérica y Sudamérica: en ambos casos, la mitad de los equipos de cada uno perdió uno o más partidos.
