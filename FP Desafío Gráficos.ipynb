##**Parte V**

Se realiza función para graficar las variables independientemente.

```python
def densidad_prob(df, variable):
  df_dropna = df[variable].dropna()
  mean_value = df_dropna.mean()
  x_min = np.min(df_dropna)
  x_max = np.max(df_dropna)
  x_axis = np.linspace(x_min, x_max, 1000)

  if isinstance(df_dropna, pd.Series) is True:
    if df_dropna.dtype != "object":
      gauss_kde = stats.gaussian_kde(df_dropna)
      z = np.reshape(gauss_kde(x_axis).T, x_axis.shape)
      plt.plot(x_axis, z, color = "c", lw = 3)
      plt.title(f"Densidad de probabilidad de {variable}")
      plt.axvline(mean_value, color = "m", linestyle = "--", lw = 1)

    else:
      raise ValueError("El objeto ingresado no tiene datos tipo integer o float, elige otra Serie.")
  else:
    raise ValueError("Los datos no son del tipo pd.Series, selecciona otros.")

densidad_prob(sample_df, "imf_pop")
```
