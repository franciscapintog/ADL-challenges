```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/worldcup2014.csv")
df
```

```python
def media(x):
  if isinstance(x, pd.Series) is True:
    if x.dtype != "object":
      x = x.dropna()
      df_mean = round(np.mean(x), 2)
    else:
      raise ValueError("El objeto no es de tipo float o integer, selecciona otro.")
  else:
    raise ValueError("Los datos ingresados no son un objeto pd.Series.")
  return df_mean

print("La media de goles a favor es:", media(df["goles_favor"]))
print("La media de goles en contra es:", media(df["goles_contra"]))
print("La media de puntos es:", media(df["puntos"]))
```

```python
def varianza(x):
  if isinstance(x, pd.Series) is True:
    if x.dtype != "object":
      x = (x - media(x))**2
      var = np.mean(x.dropna())
    else:
      raise ValueError("El objeto no es de tipo float o integer, selecciona otro.")
  else:
    raise ValueError("Los datos ingresados no son un objeto pd.Series.")
  return var

print("La varianza en la variable goles a favor es:", varianza(df["goles_favor"]))
print("La varianza en la variable goles en contra es:", varianza(df["goles_contra"]))
print("La varianza en la variable puntos es:", varianza(df["puntos"]))
```

Por otro lado, si al método np.random.choice no se le incorpora el parámetro "p", éste asume una probabilidad igual para todos los elementos del array, que en este caso son solo 2.
2. Al incorporar np.random.seed la función no cambia el array generado al ser ejecutada varias veces.

**Desafío 4**



```python
def simulate_pets_prob(size):
  np.random.seed(1)
  array2 = generate_pet(size = ((size + 1) * 2))
  array1 = ["young_pet", "old_pet"]
  pets = [] #lista con mascotas se genera vacía al iniciar

  for i in range (1, size + 1):
    #se agregarán a la lista pets cada mascota, se generan los contadores de probabilidad, y además se generan diccionarios para hacer los condicionales de cada contador
    pets.append(array1[0] + "_" + array2[i] + ", " + array1[1] + "_" + array2[i*2])
    prob_opp = 0
    prob_ypyop = 0
    prob_ypoop = 0
    dict1 = {"young_pet_perro, old_pet_gato", "young_pet_gato, old_pet_perro", "young_pet_perro, old_pet_perro"}
    dict2 = {"young_pet_perro, old_pet_perro", "young_pet_gato, old_pet_perro"}
    dict3 = {"young_pet_perro, old_pet_perro"}

    for j in pets:

      if j in dict1:
        prob_opp += 1    
    
      if j in dict2:
        prob_ypyop += 1
    
      if j in dict3:
        prob_ypoop += 1

      array3 = np.array([prob_opp, prob_ypyop, prob_ypoop])
      array3 = array3 / len(pets)

  return array3

simulate_pets_prob(1000)
```

Respecto a las consultas realizadas:

> La probabilidad de que alguna de las mascotas sea un perro es de 0.743 (o 74.3%). Por otro lado, la probabilidad de que old_pet sea un perro es de 0.492 (o 49.2%). Finalmente, la probabilidad de que ambas mascotas sean perros, es de 0.252 (o un 25.2%). A medida que crezca size, las probabilidades tenderán a 0.75, 0.5 y 0.25 respectivamente.

Respecto a las preguntas finales:

1. El escenario menos probable es el tercero (young_pet_perro y old_pet_perro simultáneamente)
2. El escenario más probable es el primero (que una o más mascotas sean perro)
3. El motivo es que en el segundo escenario existen más casos que cumplen con la condición: 3 de los 4 posibles (combinando young_pet y old_pet con perro). En cambio, en el primer escenario, solo son casos favorables la combinación young_pet_perro y old_pet_perro.
